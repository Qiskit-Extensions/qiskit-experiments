# Quantum Volume Demo

show some basic functionality of the quantum volume experiment

```python
import qiskit_experiments as qe
from qiskit import Aer

# For simulation
from qiskit.test.mock import FakeParis
backend = FakeParis()
```

## Run QV experiment

```python
qubits = 3
trials = 20

qv_exp = qe.quantum_volume.QVExperiment(qubits, trials)
expdata = qv_exp.run(backend)

# View result data
print(expdata)
```

### adding trials

in addition, returning the figures as a separate object 

```python
qv_exp.add_trials(10)
expdata2, figs = qv_exp.run(backend, experiment_data=expdata, return_figures=True)
print(expdata2)
```

```python
# figs is a list of figures. in this case, there is only 1 figure
figs[0]
```

### using another backed for simulation

```python
another_simulation_backend = Aer.get_backend('aer_simulator_matrix_product_state')
qv_exp2 = qe.quantum_volume.QVExperiment(qubits, trials, simulation_backend=another_simulation_backend)
expdata3 = qv_exp2.run(backend)
print(expdata3)
```
